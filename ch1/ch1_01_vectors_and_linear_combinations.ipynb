# 1.1. Vectors and Linear Combinations

이번 실습에서는 python의 numpy, matplotlib를 이용해서 2D 및 3D vector를 시각화 해보겠습니다.

```
import numpy as np
import matplotlib.pyplot as plt
```

2개의 요소를 가지는 vector를 2차원 공간에 시각화하기 위해 quiver라는 함수를 사용해보겠습니다. quiver라는 함수에 4개의 숫자를 연속으로 입력하였을 때 앞의 두 숫자는 vector의 시작점, 뒤의 두 숫자는 vector의 각각의 축으로부터 이동거리가 된다는 것을 확인해 볼 수 있습니다.

```
fig = plt.figure()
ax=fig.gca()
ax.quiver(0,0,5,5,angles='xy',scale_units='xy',scale=1)
ax.set_xlim([-15,15])
ax.set_ylim([-15,15])
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.grid('both')
```

3개의 요소를 가지는 vector를 3차원 공간에 시각화하기 위해 quiver함수를 사용해보겠습니다.

```
fig = plt.figure()
ax=fig.gca(projection='3d')
ax.quiver(0,0,0,5,5,5)
ax.set_xlim([-15,15])
ax.set_ylim([-15,15])
ax.set_zlim([-15,15])
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```

교재에서 정의한 2개의 2차원 vector의 linear combination이 공간상에서 어떻게 표현되는지 살펴보겠습니다.

```
v = np.transpose(np.array([1,1]))
w = np.transpose(np.array([2,3]))
print('v =',v,'w =',w)
```

두 vector를 공간상에 표현해보겠습니다.

```
plt.figure()
ax=plt.gca()
ax.quiver(0,0,v[0],v[1],angles='xy',scale_units='xy',scale=1,color='r')
ax.quiver(0,0,w[0],w[1],angles='xy',scale_units='xy',scale=1,color='g')
ax.set_xlim([-1,5])
ax.set_ylim([-1,5])
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.grid('both')
```

두 vector의 합을 공간 상에 표현해 보겠습니다.

```
plt.figure()
ax=plt.gca()
ax.quiver(0,0,v[0],v[1],angles='xy',scale_units='xy',scale=1,color='r')
ax.quiver(v[0],v[1],w[0],w[1],angles='xy',scale_units='xy',scale=1,color='g')
ax.quiver(0,0,v[0]+w[0],v[1]+w[1],angles='xy',scale_units='xy',scale=1,color='b')
ax.set_xlim([-1,5])
ax.set_ylim([-1,5])
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.grid('both')
```

```
c = 3
d = 5
r = c * v + d * w
plt.figure()
ax=plt.gca()
ax.quiver(0,0,c*v[0],c*v[1],angles='xy',scale_units='xy',scale=1,color='r')
ax.quiver(0,0,d*w[0],d*w[1],angles='xy',scale_units='xy',scale=1,color='g')
#ax.quiver(0,0,r[0],r[1],angles='xy',scale_units='xy',scale=1,color='b')
ax.set_xlim([-1,20])
ax.set_ylim([-1,20])
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.grid('both')
```

c, d의 값을 임의로 생성하기 위하여 random.randint함수를 사용합니다.

```
for n in range(10):
  c=np.random.randint(-50,50)
  d=np.random.randint(-50,50)
  print('c =',c,'d =',d)
```

100번, 1000번 이상 반복시행을 통해 임의로 생성된 c, d값에 대한 linear combination들이 공간을 어떻게 채울 수 있는 지 살펴 봅니다. vector v, w를 다르게 해서 실험을 반복해보면서 주어진 v, w의 linear combination들이 공간을 어떻게 채우는 지 시각적으로 이해합시다.

```
v = np.transpose(np.array([1,1]))
w = np.transpose(np.array([2,3]))
print('v =',v,'w =',w)

plt.figure()
ax=plt.gca()

for n in range(5):
  c=np.random.randint(-50,50)
  d=np.random.randint(-50,50)
  r = c * v + d * w

  R=np.random.randint(0,100)/(100.0)
  G=np.random.randint(0,100)/(100.0)
  B=np.random.randint(0,100)/(100.0)
  ax.quiver(0,0,r[0],r[1],angles='xy',scale_units='xy',scale=1,color=[R,G,B])

ax.set_xlim([-250,250])
ax.set_ylim([-250,250])
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.grid('both')
```

3차원 vector에 대해 같은 작업을 반복해 봅니다.

```
v = np.transpose(np.array([1, 1, 1]))
w = np.transpose(np.array([2, 3, 4]))
print('v = ',v,'w = ',w)

fig = plt.figure()
ax=fig.gca(projection='3d')
ax.quiver(0,0,0,v[0],v[1],v[2],color='r')
ax.quiver(0,0,0,w[0],w[1],w[2],color='g')
ax.set_xlim([-5,5])
ax.set_ylim([-5,5])
ax.set_zlim([-5,5])
ax.set_xlabel('x')
ax.set_ylabel('y')
```

2개의 vector v, w의 linear combination들이 3차원 공간을 어떻게 채우는지 c, d를 임의로 생성하면서 살펴봅니다. 또한, v, w를 다르게 했을 때 어떻게 변하는 지 살펴봅니다.

```
v = np.transpose(np.array([1, 1, 1]))
w = np.transpose(np.array([2, 3, 4]))
print('v = ',v,'w = ',w)

fig=plt.figure()
ax=fig.gca(projection='3d')

for n in range(1000):
  c=np.random.randint(-50,50)
  d=np.random.randint(-50,50)
  r = c * v + d * w
  R=np.random.randint(0,100)/(100.0)
  G=np.random.randint(0,100)/(100.0)
  B=np.random.randint(0,100)/(100.0)
  ax.quiver(0,0,0,r[0],r[1],r[2],color=[R,G,B])

ax.set_xlim([-250,250])
ax.set_ylim([-250,250])
ax.set_zlim([-250,250])
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.grid('both')
```

3차원에서 많은 수의 화살표로 표현을 하게 되면 시각화가 어렵기 때문에 vector의 끝을 점으로 표현하여 같은 작업을 해보겠습니다. 마찬가지로 다양한 v, w에 대해 실험을 하여 vector의 linear combination이 공간상에서 어떻게 표현되는지 이해합시다. 주어진 두 개의 vector v, w의 linear combination이 빨간색으로 표시된 (100,0,0)이라는 점을 만들어 낼 수 있을 지 고민해 봅시다.

```
v = np.transpose(np.array([1, 1, 1]))
w = np.transpose(np.array([2, 3, 4]))
print('v = ',v,'w = ',w)

fig=plt.figure()
ax=fig.gca(projection='3d')
ax.scatter3D(100,0,0,color='r',s=5)
for n in range(1000):
  c=np.random.randint(-100,100)
  d=np.random.randint(-100,100)
  r = c * v + d * w

  R=np.random.randint(0,100)/(100.0)
  G=np.random.randint(0,100)/(100.0)
  B=np.random.randint(0,100)/(100.0)
  ax.scatter3D(r[0],r[1],r[2],color='g',s=5)

ax.set_xlim([-250,250])
ax.set_ylim([-250,250])
ax.set_zlim([-250,250])
ax.set_xlabel('x')
ax.set_ylabel('y')
```

이번에는 3개의 vector의 linear combination들이 공간을 어떻게 채우는 지 살펴보겠습니다.

```
v = np.transpose(np.array([1, 1, 1]))
w = np.transpose(np.array([2, 3, 4]))
u = np.transpose(np.array([3, 4, 5]))
print('v = ',v,'w = ',w, 'u = ',u)

fig=plt.figure()
ax=fig.gca(projection='3d')

for n in range(1000):
  c=np.random.randint(-100,100)
  d=np.random.randint(-100,100)
  e=np.random.randint(-100,100)

  r = c * v + d * w + e * u

  R=np.random.randint(0,100)/(100.0)
  G=np.random.randint(0,100)/(100.0)
  B=np.random.randint(0,100)/(100.0)
  ax.scatter3D(r[0],r[1],r[2],color='g',s=5)

ax.set_xlim([-250,250])
ax.set_ylim([-250,250])
ax.set_zlim([-250,250])
ax.set_xlabel('x')
ax.set_ylabel('y')
```

만약, 세 번째 vector가 다른 두 vector의 linear combination으로 표현될 수 있으면 세 vector의 linear combination은 3차원 공간의 전체를 채울 수 없을 것입니다. 위의 실험과 아래의 실험의 결과를 비교하면서 이에 대해 생각해 봅시다.

```
v = np.transpose(np.array([2, 0, 0]))
w = np.transpose(np.array([0, 2, 0]))
u = np.transpose(np.array([0, 0, 2]))
print('v = ',v,'w = ',w, 'u = ',u)

fig=plt.figure()
ax=fig.gca(projection='3d')

for n in range(1000):
  c=np.random.randint(-100,100)
  d=np.random.randint(-100,100)
  e=np.random.randint(-100,100)

  r = c * v + d * w + e * u

  R=np.random.randint(0,100)/(100.0)
  G=np.random.randint(0,100)/(100.0)
  B=np.random.randint(0,100)/(100.0)
  ax.scatter3D(r[0],r[1],r[2],color='g',s=5)

ax.set_xlim([-250,250])
ax.set_ylim([-250,250])
ax.set_zlim([-250,250])
ax.set_xlabel('x')
ax.set_ylabel('y')
```
